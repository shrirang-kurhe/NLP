### Import the Necessary Dependencies

```python
import pandas as pd 
import re 
import string 
import nltk 
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn import metrics
```

### Read the Data 

```python
# Importing the Dataset 
twitter_df = pd.read_csv(r"C:\Users\Shree\OneDrive\Desktop\FSDS_omkar sir\Datafiles\twitter_training.csv")  # Load uploaded dataset
twitter_df.head()
```

#### Drop '2401' and Borderlands **Columns**

```python
df = twitter_df.drop(columns=['2401','Borderlands'])
df
```

#### Change Names above Columns

```python
TS_df = df.rename(columns={'Positive': 'Sentiment','im getting on borderlands and i will murder you all ,':'Review'})
TS_df
```

### Data Understanding

```python
TS_df.value_counts('Sentiment')
```

```python
TS_df.values[1]
```

### Preprocess or text cleaning

```python
ps = PorterStemmer()
def preprocess(text):
    if not isinstance(text, str):  # Check if input is not a string , If text is not a string then return Empty String("")
        return ""
    
    text = text.strip()  # Remove unwanted space
    text = re.sub("<[^>]*>", "", text)  # Remove HTML tags
    text = re.sub("[^a-zA-Z]", " ", text)  # Remove non-alphabet characters
    text = text.lower() 
    text = text.split()  # Tokenize words
    
    # Stemming
    text = [ps.stem(word) for word in text]
    return " ".join(text)
    return text
```

```python
TS_df['preprocess_review'] = TS_df['Review'].apply(preprocess)
TS_df.head()
```

```python
map_dict = {'Negative':0,
            'Positive':1,
            'Neutral':2,
            'Irrelevant':3}
TS_df['sentiment_numeric'] = TS_df.Sentiment.map(map_dict)
TS_df.head()
```

```python
TS_df.shape
```

### Perform a Train test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(TS_df.preprocess_review,
                                                TS_df.sentiment_numeric,
                                                test_size=0.2,
                                                random_state=42,
                                                stratify=TS_df.sentiment_numeric)
```

```python
print("X_train data :",x_train.shape)
print("X_test data :",x_test.shape)
print("y_train data :",y_train.shape)
print("y_test data :",y_test.shape)
```

### convert text to vector form

- **TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf = TfidfVectorizer()
tf_idf
```

#### Apply TF-IDF to Training Data

```python
x_train_tf = tf_idf.fit_transform(x_train)
print(x_train_tf.shape)
```

```python
x_test_tf = tf_idf.transform(x_test)
x_test_tf
```

### Model Creation 

 - Navie Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(x_train_tf,y_train)
```

**Prediction**

```python
y_pred = naive_bayes_classifier.predict(x_test_tf)
y_pred
```

```python
print(metrics.classification_report(y_test,y_pred,
                                    target_names=['positive','negative','Neutral','Irrelevant']))
```
