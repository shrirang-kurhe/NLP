#### Import the packages

```python
import pandas as pd 
import re 
import string 
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn import metrics
```

#### Read a Data

```python
data = pd.read_excel(r"C:\Users\Shree\Downloads\IMDB_Dataset_sample.xlsx")
data.head()
```

#### data Understand

```python
data.values[0]
```

```python
print(data.value_counts('sentiment'))
# data.value_counts('sentiment').plot
```

```python
data.value_counts('sentiment').plot(kind='bar')
```

#### Change categorial to Numerical data


```python
map_dict = {'positive':1,
            'negative':0}
data['sentiment_numeric'] = data.sentiment.map(map_dict)
data.head()
```

#### Preprocess Function

```python
ps = PorterStemmer()
corpus = set()
def preprocess(text):

    text = text.strip()
    text = re.sub('<[^>]>', '',text)
    text = re.sub('[^a-zA-Z]', ' ',text)

    text = text.lower()
    text = text.split()

 
## stemming the word for sentiment analysis do not remove the stop word
    text = [ps.stem(word) for word in text]
    text = ' '.join(text)
    return text
```

```python
data['preprocessed_review'] = data.review.apply(preprocess)
data.head()
```

```python
data.shape
```

#### Perform the Train Test Split

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(data.preprocessed_review,
                                                 data.sentiment_numeric,
                                                test_size=0.2,
                                                random_state=42,
                                                stratify=data.sentiment_numeric)
```

```python
print("X_train data :",x_train.shape)
print("X_test data :",x_test.shape)
print("y_train data :",y_train.shape)
print("y_test data :",y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf = TfidfVectorizer()
tf_idf
```

##### Apply tf-idf to training data

```python
x_train_tf = tf_idf.fit_transform(x_train)
x_train_tf
```

```python
x_test_tf = tf_idf.transform(x_test)
x_test_tf
```

```python
print(x_train_tf.shape)
print(x_test_tf.shape)
```

#### Model Creation

### **Navie bayes Classifier**

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(x_train_tf,y_train)
```

##### Prediction

```python
y_pred = naive_bayes_classifier.predict(x_test_tf)
y_pred
```

```python
print(metrics.classification_report(y_test,y_pred,
                                    target_names=['positive','negative']))
```

```python
print("Confusion matrix:")
print(metrics.confusion_matrix(y_test, y_pred))
```
